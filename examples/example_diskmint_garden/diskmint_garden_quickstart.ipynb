# DiskMINT-GARDEN Quickstart

This notebook demonstrates the beta `diskmint.garden` API. It loads the bundled surrogate models and predicts disk properties from observed continuum and $\mathrm{C^{18}O}$ fluxes.

Install the optional dependencies before running:

```bash
pip install "diskmint[garden]"
```

````python
import pandas as pd
import diskmint.garden.infer as garden
````

## One Target

Inputs are continuum flux density in mJy, integrated $\mathrm{C^{18}O}$ line flux in mJy km/s, distance in pc, stellar mass in $M_\odot$, and the 90 percent dust radius in au.

````python
result = garden.from_observations(
    flux_mm=120.0,
    flux_c18o=850.0,
    distance=140.0,
    mstar=0.8,
    rdust_90=80.0,
    band="band6",
)

{key: result[key] for key in [
    "Mgas_pred_Msun",
    "Mdust_pred_Msun",
    "gtd_pred_dimless",
    "Rc_pred_au",
    "is_outside_grid",
    "outside_features",
]}
````

## A Small Table

`from_dataframe` keeps the input columns and appends predicted quantities plus grid-domain diagnostics.

````python
table = pd.DataFrame({
    "name": ["demo-band6", "demo-band6-compact"],
    "flux_mm_mjy": [120.0, 45.0],
    "flux_c18o_mjy_kms": [850.0, 180.0],
    "distance_pc": [140.0, 160.0],
    "mstar_msun": [0.8, 0.4],
    "Rdust_90_au": [80.0, 35.0],
})

predictions = garden.from_dataframe(table, band="band6")
predictions[[
    "name",
    "Mgas_pred_Msun",
    "Mdust_pred_Msun",
    "gtd_pred_dimless",
    "Rc_pred_au",
    "is_outside_grid",
    "outside_features",
]]
````
